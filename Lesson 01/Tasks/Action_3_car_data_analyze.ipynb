```python
 import pandas as pd

# 数据加载
df = pd.read_csv('./car_complain.csv')
df

# 数据清洗，将别名合并
def f(x):
    x = x.replace('一汽-大众','一汽大众')
    return x
df['brand'] = df['brand'].apply(f)

# 拆分problem列，0代表没有，1代表存在
df = df.drop('problem', axis = 1).join(df.problem.str.get_dummies(','))
df
```

```python
# 按品牌分类，求id的计数
result_brand = df.groupby(['brand'])['id'].agg(['count'])
result_brand
```

```python
# 按车型分类，求id的计数
result_car_model = df.groupby(['car_model'])['id'].agg(['count'])
result_car_model
```

```python
# 按品牌分类，求id和car_model计数
result_brand_2 = df.groupby(['brand']).agg({'id':'count','car_model':'nunique'})
result_brand_2

# tags = df.columns[[0,2]]
# result_brand_2 = df.groupby(['brand'])[tags].agg(['nunique'])
# result_brand_2

# 索引列恢复正常列
result_brand_2.reset_index(inplace = True)
result_brand_2
```

```python
# 求各品牌下车型的平均抱怨个数
result_brand_2.eval('car_model_mean = id / car_model', inplace=True)
result_brand_2
```

```python
# 按照 car_model_mean，从大到小进行排序
result_brand_2 = result_brand_2.sort_values('car_model_mean', ascending = False)
result_brand_2
```

```python
result_brand_2.to_csv("./car_data_analyze_result.csv")
```
